```python
import requests
import time
import random
```

```python
from bs4 import BeautifulSoup as bs
```

```python
county_list=[]
prices_list=[]
age_list=[]
apartment_floor_list=[]
room_list=[]
m2_list=[]
dwellingtype_list=[]
```

```python

for i in range(1,2):
    link=f'https://www.hepsiemlak.com/zeytinburnu-satilik/daire?p7=101215-101216-101217-101218-101219-101220-101221-101222-101223-101224-101225-101226-101227-101228-101229-101230-101231-101232-101233-101234&page={i}'
    r=requests.get(link , headers={'User-Agent': 'Mozilla/5.0'})
    soup=bs(r.content,"lxml")
    fiyatlar = soup.find_all('span', {'class' : 'list-view-price'})
    konutyas=soup.find_all('span', {'class' : 'celly buildingAge'})
    kat=soup.find_all('span', {'class' : 'celly floortype'})
    odasayisi=soup.find_all('span', {'class' : 'celly houseRoomCount'})
    m2=soup.find_all('span', {'class' : 'celly squareMeter list-view-size'})
    time.sleep(random.randint(2,6))
    
    for i in range(len(fiyatlar)):
        county_list.append("Zeytinburnu")
        dwellingtype_list.append("Daire")
        room_list.append(odasayisi[i].text.replace(' ',''))
        age_list.append(konutyas[i].text)
        apartment_floor_list.append(int(kat[i].text[:-3].replace('.','')))
        m2_list.append(int(m2[i].text[:-2].replace('.','')))
        prices_list.append(int(fiyatlar[i].text[:-3].replace('.','')))
```

```python
print(county_list[2]+" "+dwellingtype_list[2]+" "+ str(prices_list[2]))
len(room_list)
```

```python
import pandas as pd
import numpy as np
data = pd.DataFrame(list(zip(county_list,dwellingtype_list,age_list,apartment_floor_list,m2_list,room_list,prices_list)),columns=["province","housing_type","age","floor","m_2","room_number","price"])
```
